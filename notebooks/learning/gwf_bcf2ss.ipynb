# 4. Plot results

## 4(a). Standard formulation simulation

````python
hds_p1a = fp.utils.binaryfile.HeadFile(Path(wksp_p1a, f"{sim_name}.hds"))
times_p1a = hds_p1a.get_times()
````

````python
fig, ax = plt.subplots(
    nrows=len(times_p1a),
    ncols=2,
    sharex=True,
    sharey=True,
    figsize=(12, 9)
)

fig.suptitle("Results of Standard Formulation", fontsize=20, fontweight="bold")
aquifer_nm = ["Upper", "Lower"]
for i, t in enumerate(times_p1a):
    for layer in [0, 1]:        
        mapview = fp.plot.PlotMapView(model=model_p1a, ax=ax[i, layer], layer=layer)
        ax[i, layer].set_title(f"{aquifer_nm[layer]} aquifer, t={t:.1f} d", fontweight="bold")
        mapview.plot_ibound()
        mapview.plot_grid()

        cb = mapview.plot_array(hds_p1a.get_data(totim=t), cmap="winter", alpha=0.6, vmin=0, vmax=140)
        plt.colorbar(cb)
        hdcn = mapview.contour_array(hds_p1a.get_data(totim=t), colors="black" if i == 0 else "white")
        plt.clabel(hdcn, fmt="%1.1f")

        if layer == 1:
            mapview.plot_bc("WEL", kper=i)
````

````python
fig, ax = plt.subplots(
    nrows=len(times_p1a),
    ncols=1,
    squeeze=True,
    figsize=(6, 10)
)
for i, t in enumerate(times_p1a):
    xsect = fp.plot.PlotCrossSection(model=model_p1a, ax=ax[i], line={"row": 7})
    ax[i].set_title(f"Row 8 Cross-Section w/ Heads, t={t:.1f} d", fontweight="bold")
    xsect.plot_ibound()
    xsect.plot_grid()
    hds_t = hds_p1a.get_data(totim=t)    
    cb = xsect.plot_array(hds_t, head=hds_t, cmap="winter", alpha=0.8, vmin=0, vmax=130)
    plt.colorbar(cb)
````
